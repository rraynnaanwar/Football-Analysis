```python
import cv2 as cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python

imagePath = r"C:\Users\rrayn\OneDrive\Desktop\Personal-Project\Football-AI\Football-Analysis\outputVideos\croppedImage.jpg"
image = cv2.imread(imagePath)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
topHalf = image[0:int(image.shape[0]/2), :]
plt.imshow(topHalf)
plt.show()
```

```python
image2D = topHalf.reshape(-1,3)
kmeans = KMeans(n_clusters=2, random_state=0).fit(image2D)
labels = kmeans.labels_
clusteredImage = labels.reshape(topHalf.shape[0], topHalf.shape[1])
plt.imshow(clusteredImage)
plt.show()
```

```python
cornerClusters = [clusteredImage[0,0], clusteredImage[0,-1], clusteredImage[-1,0], clusteredImage[-1,-1]]
nonPlayerClusters = max(set(cornerClusters), key=cornerClusters.count)
print(nonPlayerClusters)
playerClusters = 1- nonPlayerClusters
print(playerClusters)
```

```python
kmeans.cluster_centers_[playerClusters]
```
